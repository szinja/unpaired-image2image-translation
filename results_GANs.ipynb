### Input | Baseline Output | CLIP Output

```python
import glob
from PIL import Image
from IPython.display import display, HTML

# Define the path to your qualitative comparison images
image_folder = "/content/unpaired-image2image-translation/evaluation_results_A2B/qualitative_comparison"

# Find all PNG images in the folder
# Use sorted() to ensure a consistent order, adjust if needed
image_files = sorted(glob.glob(f"{image_folder}/*.png"))

print(f"Found {len(image_files)} images in {image_folder}")

# Display each image
num_to_display = 5
images_to_show = image_files[:num_to_display]
if not images_to_show:
    print("No images selected in the specified directory.")
else:
    display(HTML(f"<h2>Displaying First {len(images_to_show)} Qualitative Comparison Results (Input | Baseline | CLIP)</h2>"))
    for img_path in images_to_show:
        try:
          # Displaying my final side by side model images
            display(Image.open(img_path))
        except Exception as e:
            print(f"Error displaying image {img_path}: {e}")
```
